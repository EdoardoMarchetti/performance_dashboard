```python
import distinctipy
import os 
import os.path as osp
import json
```

##### Metrics colors

```python
with open(osp.join('..','glossaries', 'metrics.json')) as f:
        metrics =  json.load(f)

colors = distinctipy.get_colors(len(metrics), pastel_factor=0.8)
for c,m in zip(colors,metrics):
        m['color'] = c

```

```python
distinctipy.color_swatch(colors)
```

```python
with open(osp.join('..','glossaries', 'metrics.json'), 'w') as f:
    json.dump(metrics, f) 
```
